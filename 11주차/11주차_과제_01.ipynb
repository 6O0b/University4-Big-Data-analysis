### import 

```python
import torch
```

### data 

```python
ones= torch.ones(5)
x = torch.tensor([11.0,12.0,13.0,14.0,15.0])
X = torch.vstack([ones,x]).T
y = torch.tensor([17.7,18.5,21.2,23.6,24.2])

W = torch.tensor([3.0,3.0]) 

u = X@W 
v = y-u 
```

### 실행

```python
v
```

```python
_v= torch.tensor([-18.3000, -20.5000, -20.8000, -21.4000, -23.8000],requires_grad=True)
```

```python
_loss = _v.T @ _v 
```

```python
_loss.backward() 
```

```python
_v.grad.data, v
```
